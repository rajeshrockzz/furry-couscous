```
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```
type(cancer)
```

```
cancer.data #X
```

```
cancer.target # y
```

```
cancer.target_names
```

```
cancer.feature_names
```

```
print(cancer.DESCR)
```

```
X = cancer.data
y = cancer.target
```

```
X.shape
```

```
y.shape
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 5, test_size = 0.2,stratify= y)
```

```
X_train.shape,X_test.shape
```

```
from sklearn.tree import DecisionTreeClassifier
```

https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html?highlight=decisiontreeclassifier#sklearn.tree.DecisionTreeClassifier

```
tree_clf = DecisionTreeClassifier(random_state=100)
```

```
tree_clf.fit(X_train,y_train)
```

```
y_pred = tree_clf.predict(X_test)
```

```
from sklearn.metrics import accuracy_score
acc_score = accuracy_score(y_test,y_pred)
print(acc_score)
```

```
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```
print(confusion_matrix(y_test,y_pred))
```

```
print(classification_report(y_test,y_pred))
```

```
tree_clf.get_depth()
```

```
from sklearn.tree import export_graphviz
import pydotplus as pdot
from IPython.display import Image
```

```
export_graphviz(tree_clf, out_file = "decision_tree_1.odt",
                feature_names = cancer.feature_names,
                class_names = cancer.target_names,
                filled = True)
```

```
decision_tree_graph = pdot.graphviz.graph_from_dot_file("decision_tree_1.odt")
```

```
decision_tree_graph.write_jpg("decision_tree_1.jpg")
```

```
Image("decision_tree_1.jpg")
```

# Entropy as impurity index

```
tree_clf = DecisionTreeClassifier(criterion = "gini", max_depth = 3,random_state=100)
```

```
tree_clf.fit(X_train,y_train)
```

```
tree_clf.get_depth()
```

```
y_pred = tree_clf.predict(X_test)
```

```
accuracy_score(y_test,y_pred)
```

```
confusion_matrix(y_test,y_pred)
```

```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
sns.heatmap(confusion_matrix(y_test,y_pred), fmt='', cmap='Blues')
```

```
print(classification_report(y_test,y_pred))
```

```
export_graphviz(tree_clf, out_file = "decision_tree_2.odt",
                feature_names = cancer.feature_names,
                class_names = cancer.target_names,
                filled = True)
```

```
decision_tree_graph = pdot.graphviz.graph_from_dot_file("decision_tree_2.odt")
decision_tree_graph.write_jpg("decision_tree_2.jpg")
Image("decision_tree_2.jpg")
```

```
# GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```
clf_tree = DecisionTreeClassifier()
```

```
grid_cv = GridSearchCV(clf_tree, #instance of ML Model
                       param_grid= {"criterion" : ["gini","entropy"] , "max_depth" : range(2,30) , "splitter" : ["best","random"] } ,
                       cv = 10, 
                       scoring="roc_auc" )
```

```
grid_cv.fit(X_train,y_train)
```

```
grid_cv.best_params_
```
